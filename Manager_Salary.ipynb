### MEMBUAT KOLOM SALARY BARU DENGAN SATUAN CURRENCY YANG SAMA YAITU USD

```python
# Nilai tukar terhadap USD BERDASAKAN KURS NILAI MATA UANG HARI INI
konversi_mata_uang = {'USD': 1, 'ZAR': 0.05311, 'SEK': 0.09571, 'HKD': 0.12832, 'NOK': 0.09348, 'BRL': 0.20445, 'DKK': 0.14683,
                      'TTD': 0.14715, 'MXN': 0.05838, 'CZK': 0.04485, 'BDT': 0.00907, 'PHP': 0.01804, 'PLN': 0.25056, 'TRY': 0.03465 ,
                      'CNY': 0.13987, 'ILS': 0.26757, 'JPY': 0.00669, 'TWD': 0.03159, 'SGD': 0.74666, 'KRW': 0.00077, 'THB': 0.02831,
                      'IDR': 0.00006, 'NZD': 0.60835, 'LKR': 0.00305, 'SAR': 0.26679, 'NTD': 0.03159, 'HRK': 0.14200, 'NGN': 0.00122, 'COP': 0.00025,
                      'GBP': 1.26060, 'CAD': 0.73457,'EUR':1.09450 ,'AUD': 0.65900,'INR':0.01200  ,'ARS':0.00280 ,'CHF':1.13372 ,'MYR': 0.21347 }

# Fungsi untuk konversi nilai mata uang ke USD
def convert_to_usd(row):
    if row['currency'] == 'USD':
        return row['annual_salary']
    elif row['currency'] in konversi_mata_uang:
        return row['annual_salary'] * konversi_mata_uang[row['currency']]
    else:
        return None  # Nilai 'UNKNOWN' atau mata uang lain yang tidak diketahui

# Tambahkan kolom baru 'amount_usd' dengan nilai konversi
data['sallary_USD'] = data.apply(convert_to_usd, axis=1)

```

## NORMALISASI

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_normalized = pd.DataFrame(scaler.fit_transform(df.drop('sallary_USD', axis=1)))
df_normalized['sallary_USD'] = df['sallary_USD']

df_normalized.head()
```

## SPLIT DATA

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_normalized.drop('sallary_USD', axis=1),\
                                                    df_normalized['sallary_USD'], test_size=0.2, random_state=1911)
```

Membagi data training dan testing menjadil 80:20

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## MODELING LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# Membuat model regresi linear berganda
model = LinearRegression()

# Melatih model menggunakan data pelatihan
model.fit(X_train, y_train)

# Membuat prediksi menggunakan data pengujian
y_pred = model.predict(X_test)

for i in range(5):
  print(f"groundtruth: {y_test.iloc[i]}")
  print(f"prediction : {y_pred[i]}")
  print("==="*8)

# # Evaluasi performa model
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred, squared=False))

# # Plot hasil prediksi
# plt.scatter(y_test, y_pred)
# plt.xlabel('Nilai Aktual')
# plt.ylabel('Prediksi')
# plt.title('Perbandingan Nilai Aktual dan Prediksi')
# plt.show()
```

## SUPPORT VECTOR REGRESSION

```python
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error

svr_model = SVR(kernel='linear', C=1.0, epsilon=0.2)

# Latih model pada set pelatihan yang telah diubah skala
svr_model.fit(X_train, y_train)

# Prediksi pada set pengujian
y_pred = svr_model.predict(X_test)


for i in range(5):
  print(f"groundtruth: {y_test.iloc[i]}")
  print(f"prediction : {y_pred[i]}")
  print("==="*8)
# Evaluasi performa model
mse = mean_squared_error(y_test, y_pred)
print(f'Mean Squared Error: {mse}')
```
