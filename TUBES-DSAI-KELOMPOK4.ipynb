2. Strategy experimen mennggunakan Grid Search
Pencarian grid telah menjadi standar optimisasi parameter yang ada dalam pembelajaran mesin, yang merupakan pencarian menyeluruh melalui subset yang didefinisikan secara manual dari ruang hiperparameter algoritma pembelajaran.Karena kemudahannya dalam pelaksanaan dan paralelisasi serta keandalannya dalam ruang dimensi rendah, pencarian grid tetap menjadi metode terdepan dalam optimisasi hiperparameter.

Dalam penelitian yang dilakukan oleh (Belete & Huchaiah, 2021) bahwa diterapkan grid search untuk menemukan hypermeter tuning terbaik . 3. Skema validasi menggunakan K-fold cross validation dan StratifiedKFold dengan nilai k= 5.
K-fold cross validation digunakan untuk melakukan pembagian data berdasarkan jumlah fold. Pembagian dilakukan secara acak dengan komposisi tertentu. Pembagian data dalam k-fold biasanya menggunakan nilai ‘k’ atau data akan di-fold sejumlah ‘k’ yang telah diinisiasi di awal. Banyaknya nilai ‘k’ menentukan berapa banyak kombinasi data pelatihan dan data pengujian yang akan diterapkan pada model. Pada setiap fold, akan diberikan skor akurasi yang menunjukkan hasil pelatihan model terhadap data-data yang sudah dibagi tadi. Hasil fold dengan akurasi tertinggi akan dipilih sebagai model yang akan digunakan ketika pengujian pada data baru yang tidak pernah dilihat oleh mesin sebelumnya. 
K-fold akan cocok digunakan ketika data yang dimiliki tidak banyak, sehingga memerlukan bantuan k-fold untuk melakukan pelatihan awal terhadap data-data acak melalui k-fold. Pelatihan model terhadap data yang dibagi melalui k-fold, membantu kinerja model untuk lebih bisa tergeneralisasi pada data baru nantinya. Setiap fold akan digunakan sebagai data uji dan k-1 fold yang lainnya akan digunakan sebagai data pelatihan. Kemudian hasil fold yang memberikan akurasi terbaik, akan dipakai kembali modelnya untuk melalui pelatihan lagi dengan keseluruhan data latih. Proses tersebut bertujuan untuk memperkuat model yang telah melalui pelatihan dengan data k-fold, sehingga model dapat lebih baik dalam menggeneralisasi pola dari data untuk diujikan pada data baru.

Metode yang digunakan untuk membagi dataset dengan K-Fold Cross Validation, berdasarkan penelitian (Saputro, et al, 2020) didapatkan bahwa K-Fold Cross Validation dapat membagi data dengan baik dan menghindari terjadinya overfitting terhadap model. K-Fold cross validation juga dapat mengevaluasi kinerja model dengan menampilkan akurasi yang terbesar diantara nilai fold masukan. Penelitian berdasarkan (Hastie, et al, 2009) mendapati penggunaan k=5 ataupun k=10 menjadi masukan yang ideal dan menghasilkan akurasi yang cukup optimal pada nilai ‘k’ tersebut. Eksperimen ini menggunakan k-fold cross validation untuk menghindari terjadinya overfitting, yaitu kondisi ketika akurasi pelatihan memiliki nilai yang tinggi tetapi ketika dilakukan pengujian dengan data baru, akurasi mengalami penurunan yang sangat jauh. Hal ini menunjukkan bahwa model tidak bisa tergeneralisasi prediksinya pada data baru. Untuk menghindari hal tersebut, eksperimen akan menggunakan k-fold cross validation sejumlah 5 k, sehingga tidak akan mengalami komputasi yang terlalu lama mengingat kurangnya ketersediaan sumber daya komputasi. Nilai k=5 juga termasuk nilai yang paling umum digunakan dalam k-fold cross validation, sehingga aman jika digunakan dalam eksperimen ini. Stratified K-Fold digunakan untuk melakukan validasi, dengan mempertahankan proporsi kelas saat membagi data.

4. Oversampling dan Undersampling
Kelas customer churn antara pelanggan yang beralih dengan yang tidak, mengalami jumlah yang tidak seimbang. Perbedaannya sebanyak 27% untuk 'beralih' dan 73% untuk 'tidak beralih'. Oversampling akan melakukan pengacakan secara otomatis untuk membuat data baru pada kelas minoritas, sehingga jumlah kelasnya akan menyamakan kelas mayoritas. Sedangkan undersampling akan bekerja sebaliknya, dengan menyamakan kelas mayoritas ke jumlah kelas minoritas. Oversampling yang dilakukan akan menggunakan smote dan randomoversampler, sedangkan untuk undersampling nya akan menggunakan randomundersampler.

Eksperimen terhadap model dengan pemilihan fitur-fitur berdasarkan korelasinya, tidak memberikan hasil yang cukup signifikan dibandingkan eksperimen lainnya. Penggunaan fitur yang lebih sedikit memang menambah efektifitas kinerja sebuah model, tetapi tidak menutup kemungkinan bahwa pemangkasan fitur malah menyebabkan model tidak bekerja lebih baik. Terdapat beberapa cara untuk melakukan pemangkasan atau pengurangan fitur yang digunakan, beberapa yang dilakukan dalam percobaan ini ialah dengan memilih fitur berdasarkan korelasi, memilih berdasarkan information gain, dan memangkas fitur menggunakan PCA.

Penggunaan hyperparameter secara umum membantu model machine learning, untuk bisa mengatur cara kerjanya sehingga sesuai dengan kasus yang sedang ditangani. Kombinasi hyperparameter yang cocok biasanya harus melalui 'trial and error' atau bisa juga memanfaatkan pencarian otomatis berupa grid search atau random search (jurnal). Nilai yang tertera pada setiap parameter model ada yang bergantung pada jumlah kelas, kompleksitas kasus, dan juga referensi penelitian terdahulu (melihat nilai-nilai yang paling efektif digunakan). Hyperparameter membantu mengatasi overfitting atau mampu melakukan generalisasi terhadap pengujian setelah model terlatih. Namun tidak menutup kemungkinan, hyperparameter yang telah diimplementasikan dan dirasa kurang tepat, juga bisa menyebabkan underfitting. Model yang memberikan hasil tidak bagus meskipun sudah menggunakan hyperparameter, bisa disebabkan oleh beberapa hal, salah satunya adalah interaksi antar hyperparameter. Nilai yang ditingkatkan pada salah satu hyperparameter sebuah model, mungkin mengharuskan penyesuaian lagi pada hyperparameter lain untuk mencegah overfitting.

Support Vector Machine atau SVM merupakan model yang cocok digunakan salah satunya untuk klasifikasi. Hyperparameter yang digunakan pada SVM biasanya ada C, gamma, dan kernel. Hyperparameter 'C' adalah parameter yang mengontrol trade-off antara mendapatkan decision boundary yang benar-benar memisahkan data pelatihan ataupun menciptakan decision boundary yang lebih umum. Nilai C yang tinggi cenderung memberikan model SVM yang lebih kompleks dan berpotensi overfitting pada data pelatihan. Nilai C yang rendah, sebaliknya, dapat menghasilkan decision boundary yang lebih sederhana dan memiliki kemungkinan underfitting. Gamma mengontrol bentuk dari decision boundary. Nilai gamma yang tinggi akan menghasilkan decision boundary yang lebih kompleks dan dapat menyesuaikan dengan data pelatihan dengan baik, tetapi dapat menyebabkan overfitting. Kernel adalah fungsi matematika yang digunakan untuk mengubah ruang fitur asli data menjadi ruang fitur yang lebih tinggi dimensi. Fungsi ini memungkinkan SVM untuk menangani data yang tidak linear. Gambar persebaran titik antara tenure dan monthly charge terhadap kelas churn salah satunya menunjukkan persebaran pada kasus ini yang tidak linear. Penting untuk bisa memutuskan nilai yang tepat pada C, gamma, dan kernel serta melakukan penyesuaian nilai antara ketiganya, yang perlu dilakukan agar terjadi overfitting. Pemilihan kombinasi nilai c, gamma, dan kernel yang digunakan berdasarkan nilai hyperparameter tersebut yang umum atau sering dipakai dalam beberapa penelitian terkait. Ketiga hyperparameter digunakan dalam kasus customer churn agar model bisa menyesuaikan kondisi data yang tidak linear dengan adanya bantuan dari kernel, selain itu nilai c dan gamma yang dapat membantu model menyesuaikan titik pemisahan data atau kelas customer churn. Nilai c dan gamma membantu titik data maupun garis pemisah kelas agar semakin mendekati garis pemisah.

Naive Bayes merupakan algoritma probabilistik yang digunakan untuk klasifikasi. Algoritma ini didasarkan pada teorema bayes. Beberapa hyperparameter yang digunakan dalam naive bayes yaitu nilai alpha, fit_prior, dan class_prior. Nilai alpha membantu model menghindari adanya nilai probabilitas yang 0. Semakin besar nilai alpha maka semakin banyak smoothing yang diberikan. Istilah smoothing merujuk pada penyetaraan nilai sehingga tidak ada probabilitas 0 pada hasil model naive bayes. Namun, semakin tinggi nilai alpha bisa saja menyebabkan underfitting, karena model menjadi kurang kompleks mengingat nilai asli probabilitas yang diubah-ubah untuk menyesuaikan nilai yang 0. Fit_prior ialah parameter boolean yang menentukan apakah akan menyesuaikan prior kelas atau menggunakan prior yang disesuaikan, biasanya akan ditentukan nilai fit_prior berupa True atau False.
